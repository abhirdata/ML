```python
u1=pd.DataFrame(u.groupby(["num_published_lectures"])["num_subscribers"].sum())
```

```python
u1['Lectures']=u1.index
```

```python
u1
```

```python
plt.bar(u1['Lectures'],u1['num_subscribers'])
plt.xlabel('Lectures')
plt.ylabel('Number of Subs')
```

```python
"""CONCLUSION FROM THIS GRAPH:
   maximum number of subscribers lie between the range 0 to 100 published lectures
   so number of published of lectures exceeding 100 is found to be tedious thus less subscriptions
   0-100 lectures can be optimum number of lectures for a particular course"""
```
