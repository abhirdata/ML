```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
```

```python
objLis = c.drop(['price'],axis=1).select_dtypes(exclude = "object").columns
```

```python
for feat in objLis:
    c[feat] = s.fit_transform(c[feat].to_numpy().reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
x=c.drop(['price'],axis=1)
y=c['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
plt.figure(figsize=(18,16))
cor=x_train.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
objList = x_train.select_dtypes(include = "object").columns
print (objList)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    x_train[feat] = le.fit_transform(x_train[feat].astype(str))

print (x_train.info())
```

```python
objList1 = x_test.select_dtypes(include = "object").columns
print (objList1)
```

```python
for feat in objList1:
    x_test[feat] = le.fit_transform(x_test[feat].astype(str))

print (x_test.info())
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(x_train, 0.75)
len(set(corr_features))
```

```python
corr_features
```

```python
x_train=x_train.drop(corr_features,axis=1)
x_test=x_test.drop(corr_features,axis=1)
```

```python
"""Car Company also affects the rates"""
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)
print("Predicted Prices: {}".format(y_pred_lr))
```

```python
check=pd.DataFrame({"Actual":y_test,"Predicted":y_pred_lr})
check
```

```python
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_lr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_lr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_lr))
print('R squared: {:.2f}'.format(lr.score(x_train,y_train)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```
